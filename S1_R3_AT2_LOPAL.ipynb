```python
# Exercício 5

# Nome
while True:
    nome = input("Digite o nome: ")

    if len(nome) > 3:
        break
    else:
        print("Nome deve ter mais de 3 caracteres.")

# Idade
while True:
    idade = int(input("Digite a idade: "))

    if 0 <= idade <= 150:
        break
    else:
        print("Idade inválida.")

# Salário
while True:
    salario = float(input("Digite o salário: "))

    if salario > 0:
        break
    else:
        print("Salário deve ser maior que zero.")

# Sexo
while True:
    sexo = input("Digite o sexo (f/m): ").lower()

    if sexo == "f" or sexo == "m":
        break
    else:
        print("Sexo inválido.")

# Estado Civil
while True:
    estado_civil = input("Digite o estado civil (s/c/v/d): ").lower()

    if estado_civil in ["s", "c", "v", "d"]:
        break
    else:
        print("Estado civil inválido.")

print("\nDados cadastrados com sucesso!")
```

```python
# Exercício 6

numero = int(input("Digite um número inteiro: "))
if numero > 1:
    primo = True
    for i in range(2, numero):
        if numero % i == 0:
            primo = False
            break
    if primo:
        print(numero, "é primo.")
    else:
        print(numero, "não é primo.")
else:
    print("Número inválido para primo.")
```

```python
# Exercício 7

numero = int(input("Digite um número: "))
fatorial = 1
for i in range(1, numero + 1):
    fatorial = fatorial * i
print("Fatorial =", fatorial)
```

```python
# Exercício 8

L = [5, 7, 2, 9, 4, 1, 3]

# a) tamanho da lista
print("Tamanho da lista:", len(L))
# b) maior valor
print("Maior valor:", max(L))
# c) menor valor
print("Menor valor:", min(L))
# d) soma dos elementos
print("Soma dos elementos:", sum(L))
# e) ordem crescente
print("Ordem crescente:", sorted(L))
# f) ordem decrescente
print("Ordem decrescente:", sorted(L, reverse=True))
```

```python
# Exercício 10

senha_correta = "676767"
senha = ""
while senha != senha_correta:
    senha = input("Digite a senha: ")
    if senha != senha_correta:
        print("Senha incorreta!")
print("Acesso liberado")
```
